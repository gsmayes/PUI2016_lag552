```python
from __future__ import print_function
import pylab as pl
from scipy.optimize import curve_fit, minimize
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
    
%pylab inline
```

# Normal Distribution

```python
normsize = (2000 / (np.array(range(1, 100)))).astype(int)
print (normsize, normsize.shape)
```

```python
normmean = 100
```

```python
md = {}
collective = []
```

```python
norm_dist = ['norm']
#distributions = ['norm']
```

```python
df = normmean - 1
```

```python
md['norm'] = {} 

for n in normsize:
    md['norm'][n] = np.random.normal(loc=normmean, scale=10, size = n)

md['norm']['means'] = {}
axnorm_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['norm']:
    if not type(nn) == str:
        md['norm']['means'][nn] = md['norm'][nn].mean()
        axnorm_mu_n.plot(nn, md['norm']['means'][nn], 'o')
        axnorm_mu_n.set_xlabel('sample size', fontsize=18)
        axnorm_mu_n.set_ylabel('sample mean', fontsize=18)
        axnorm_mu_n.set_title('Normal', fontsize=18)
        axnorm_mu_n.plot([min(normsize), max(normsize)], [normmean, normmean], 'k')
        collective.append(md['norm']['means'][nn])
#print (collective)
```

## Figure 1: Normal Distribution
This figure of a normal distribution demonstrates the Central Limit Theorem (and law of large numbers), which states that the larger the sample size, the closer the distribution of sample means will be to the true population mean, and that it will more closely approach a Gaussian distribution. As seen here, the larger the sample size along the x-axis, the closer the sample means are to the true parameter of 100. Conversely, the smaller sample sizes show greater variance in their means around 100. The bell-shaped curve of a Gaussian (or normal) distribution is also apparent, with a larger number of samples having means closer to 100 in the center.

## Figure 3: Poisson Distribution

As seen here, the larger the sample size along the x-axis, the closer the sample means are to the true parameter of 100. Conversely, the smaller sample sizes show greater variance in their means around 100. The bell-shaped curve of a Gaussian (or normal) distribution is also apparent, with a larger number of samples having means closer to 100 in the center). As a note, poisson distributions show the likelihood of an event happening over a certain time interval. The greater number of poisson samples taken, the closer it appears to a normal distribution (according to the CLT). As expected with a Poisson distribution, however, the distribution is slightly right-skewed, since the number of samples isn’t high enough to create a perfectly normal distribution.

## Figure 4: Binomial Distribution

Unlike the examples before, however, the mean of a binomial distribution is the product of the sample number and the probability, which I've set to 0.50. This was done because binomial distributions are used for yes/no situations, where the probability is usually 50% for either answer. for this reason I chose 200 instead of 100 for the 'mean' in this case, so that the distribution would actually be centered around 100 (and reflect the mean parameter for the rest of the distributions in the assignment. As seen here, the larger the sample size along the x-axis, the closer the sample means are to the true parameter of 100. Conversely, the smaller sample sizes show greater variance in their means around 100 The bell-shaped curve of a Gaussian (or normal) distribution is also apparent, with a larger number of samples having means closer to 100 in the center).
